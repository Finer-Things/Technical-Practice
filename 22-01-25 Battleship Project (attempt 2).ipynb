```python
def ship_length(data_list):
    length_count = 0
    for num in data_list:
        if num == 0:
            break
        else:
            length_count += 1
    return length_count
```

```python
def list_battle_ships(matrix):
    n = len(matrix)
    horizontal_ship_dict = {} #dictionary of horizontal ships based on initial location square
    vertical_ship_dict = {} #dictionary of vertical ships based on initial location square
    
    for i, row in enumerate(matrix):
        for j, entry in enumerate(row):
            if i > 0 and matrix[i-1][j] == 1:
                pass
            elif j > 0 and matrix[i][j-1] == 1:
                pass
            else: 
                horizontal_ship_dict[(i,j)] = ship_length(matrix[i][j:])
    
    for i, row in enumerate(matrix):
        for j, entry in enumerate(row):
            vertical_ship_dict[(i,j)] = ship_length([matrix[k][j] for k in range(i,n)])
    
    
    
    return horizontal_ship_dict, vertical_ship_dict
```

```python
def ship_counter(matrix):
    n = len(matrix)
    for row in matrix: #Makes sure all rows are the correct length
        if len(row) != n:
            return "Error: The entry  at " + len(row) + " is not the correct length."
        
    hsd, vsd = list_battle_ships(matrix)
    ship_dict = {}
    ones_list = [(i,j) for i in range(n) for j in range(n) if hsd.get((i,j)) == 1 and vsd.get((i,j)) == 1]
    twos_list = [(i,j) for i in range(n) for j in range(n) if hsd.get((i,j)) == 2 or vsd.get((i,j)) == 2]
    threes_list = [(i,j) for i in range(n) for j in range(n) if hsd.get((i,j)) == 3 or vsd.get((i,j)) == 3]
    fours_list = [(i,j) for i in range(n) for j in range(n) if hsd.get((i,j)) == 4 or vsd.get((i,j)) == 4]
    return len(ones_list), len(twos_list), len(threes_list), len(fours_list), ones_list, twos_list, threes_list, fours_list, hsd, vsd
                
```

```python
def touching_ships_checker(matrix):
    n = len(matrix)
    #for loops to make sure that there are not adjacent 1's in two L.I. directions. 
    for i, row in enumerate(matrix):
        for j, entry in enumerate(row):
            if entry == 1:
                if i != 0 and j != 0:
                    if matrix[i+1][j] == 1 and matrix[i][j+1] == 1: #below and right
                        return False #Outputs False to signal two touching ships
                if i != 0 and j != n-1:
                    if matrix[i+1][j] == 1 and matrix[i][j-1] == 1: #below and left
                        return False
                if i != n-1 and j != 0:
                    if matrix[i-1][j] == 1 and matrix[i][j+1] == 1: #above and right
                        return False 
                if i != n-1 and j != n-1:
                    if matrix[i-1][j] == 1 and matrix[i][j-1] == 1: #above and left
                        return False
    return True
                
                
            
                
    
```

```python
def battleship_checker_function(matrix):
    n = len(matrix)
    if touching_ships_checker(matrix) == False:
        return "Alpha"
    ship_length_info = list(ship_counter(matrix)) #lengths: 1s, 2s, 3s, 4s, Lists: 1s, 2s, 3s, 4s, Dicts: hsd, vsd
    #ship count
    for r in range(4):
        if ship_length_info[r] != 4-r:
            return "Beta"
    #longer ships?
    for num in ship_length_info[-2].values:
        if num > 4:
            return "Gamma"
    for num in ship_length_info[-1].values:
        if num > 4:
            return "Delta"
    return True
    
    
```
